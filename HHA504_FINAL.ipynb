```python
df_Hospital['provider_id'] = df_Hospital['provider_id'].astype(str)
df_Outpatient['provider_id'] = df_Outpatient['provider_id'].astype(str)
```

```python
# df_merged = df1.merge(df2, how='left', left_on='xx', right_on='xx')
```

```python
df_merged = df_Outpatient.merge(df_Hospital, how='left', left_on='provider_id', right_on='provider_id')
```

```python
df_merged.sample(10)
```

```python
df_merged
```

```python
# remove any NaN from hospital_name
df_merged_clean = df_merged[df_merged['hospital_name'].notna()]
```

```python
df_merged_clean
```

```python
df_merged_clean_SB = df_merged_clean[df_merged_clean['hospital_name'] == 'SUNY/STONY BROOK UNIVERSITY HOSPITAL']
df_merged_clean_SB
```

```python
df_merged_clean_GH = df_merged_clean[df_merged_clean['hospital_name'] == 'ADVENTHEALTH GORDON']
df_merged_clean_GH
```

```python
df_comparison = pd.concat([df_merged_clean_GH, df_merged_clean_SB])
df_comparison
```

```python
# pd.pivot(index='foo', columns='bar', values='baz')
df_comparison.pivot(index='provider_name', columns='apc', values='average_estimated_submitted_charges')
```
